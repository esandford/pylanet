```python
from __future__ import division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pylanet

import astropy.units as u
from astropy.constants import G

% matplotlib inline
```

```python
kep10 = pylanet.data.getData(10.01,"./downloadedData/",cadence="short")
```

```python
print np.shape(kep10.times)
print np.shape(kep10.cadencenos)

print np.shape(np.hstack(kep10.times))
print np.shape(np.hstack(kep10.sap_fluxs))
#help(kep10)
```

```python
colorcycle = ['b','r','g','m','k','c','y']
```

```python
fig = plt.figure(figsize=(18,6))

for i in range(0,len(kep10.times)):
    plt.plot(kep10.times[i],kep10.sap_fluxs[i],color=colorcycle[i%7],marker='.',ls='None',alpha=1)
plt.xlim(240,250)
plt.ylim(42000,43500)
plt.xlabel("time [BJD - 2454833]")
plt.ylabel("SAP flux [e-/s]")
#plt.show()
plt.savefig("./koi10.01.png",fmt="png")
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(np.hstack(kep10.times),np.hstack(kep10.sap_fluxs),'k.',alpha=0.25)
plt.xlim(200,400)
plt.show()
```
